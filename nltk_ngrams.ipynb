```python
import nltk 
nltk.download("book")
from nltk.book import *
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
# stopwords
stop_words = stopwords.words('english')
# Eliminado de stopwords y conversión a lowercase
text1 = [word.lower() for word in text1 if word not in stop_words]
```

## Bi-gramas

```python
# Generación de bigramas
bigramas = list(bigrams(text1))
# Frecuencia de bigramas
fdist = FreqDist(bigramas)
```

```python
# Bigramas más frecuentes
fdist.most_common(10)
```

```python
fdist.plot(10)
```

## Filtrado de bigramas

```python
# Filtrado de bigramas con más de 2 caracteres
filtered_bigrams = [bi for bi in bigramas if len(bi[0]) > 2 and len(bi[1]) > 2]
```

```python
filtered_dist = FreqDist(filtered_bigrams)
filtered_dist.plot(10)
```

## Tri-gramas

```python
from nltk.util import ngrams

trigrams = list(ngrams(text1, 3))
trigrams = [tri for tri in trigrams if len(tri[0])>2 and len(tri[1])>3 and len(tri[2])>2]
fdist = FreqDist(trigrams)
fdist.plot(10)
```
